```python
import matplotlib.pyplot as plt
import pandas as pd

# Datos corregidos
results = pd.DataFrame({
    "var_scale": [0.10, 0.25, 0.40, 0.60, 0.80, 1.00],
    "accuracy": [0.999, 0.987, 0.956, 0.919, 0.838, 0.802],
    "sensitivity": [0.998, 0.978, 0.936, 0.912, 0.820, 0.826],
    "specificity": [1.000, 0.996, 0.976, 0.926, 0.856, 0.778],
    "f1": [0.998999, 0.986882, 0.955102, 0.918429, 0.835031, 0.806641]
})

# Crear figura
plt.figure(figsize=(7,5))

# Definir marcadores y métricas
markers = ['o', 's', '^', 'x']  
metrics = ["accuracy", "sensitivity", "specificity", "f1"]

for i, metric in enumerate(metrics):
    # Desplazamiento en X para evitar sobreposición
    offset = (i-1.5)*0.01
    plt.plot([v+offset for v in results["var_scale"]], results[metric],
             marker=markers[i], markersize=7,
             linewidth=2, label=metric.capitalize())

# Escala logarítmica en eje X
plt.xscale("log")

# Configuración de ejes y título
plt.ylim(0.75, 1.02)
plt.xlabel("Noise scale (var_scale)")
plt.ylabel("Performance")
plt.title("Robustness of TEKTE-Net under increasing noise")

# Grid y leyenda
plt.grid(True, which="both", linestyle="--", linewidth=0.7, alpha=0.7)
plt.legend()

# Ajustar diseño
plt.tight_layout()

# Guardar en Kaggle working
plt.savefig("/kaggle/working/fig_robustness.png", dpi=300)
plt.show()

print(" Figura guardada en /kaggle/working/fig_robustness.png")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Datos corregidos
results = pd.DataFrame({
    "var_scale": [0.10, 0.25, 0.40, 0.60, 0.80, 1.00],
    "accuracy": [0.999, 0.987, 0.956, 0.919, 0.838, 0.802],
    "sensitivity": [0.998, 0.978, 0.936, 0.912, 0.820, 0.826],
    "specificity": [1.000, 0.996, 0.976, 0.926, 0.856, 0.778],
    "f1": [0.998999, 0.986882, 0.955102, 0.918429, 0.835031, 0.806641]
})

# Pasar métricas a porcentaje
results_pct = results.copy()
for metric in ["accuracy", "sensitivity", "specificity", "f1"]:
    results_pct[metric] = results_pct[metric] * 100

# Crear figura
plt.figure(figsize=(7,5))

# Definir marcadores y métricas
markers = ['o', 's', '^', 'x']  
metrics = ["accuracy", "sensitivity", "specificity", "f1"]

for i, metric in enumerate(metrics):
    # Agregar un pequeño desplazamiento en X para evitar solapamiento
    offset = (i-1.5)*0.01
    plt.plot([v+offset for v in results_pct["var_scale"]], results_pct[metric],
             marker=markers[i], markersize=7,
             linewidth=2, label=metric.capitalize())

# Eje X lineal
plt.xscale("linear")

# Configuración de ejes y título
plt.ylim(70, 102)
plt.xlabel("Noise scale")
plt.ylabel("Performance (%)")
# plt.title("Robustness of TEKTE-Net under increasing noise")

# Grid y leyenda
# plt.grid(True, linestyle="--", linewidth=0.7, alpha=0.7)
plt.legend()

# Ajustar diseño
plt.tight_layout()

# Guardar en Kaggle working
plt.savefig("/kaggle/working/fig_robustness_pct.png", dpi=300)
plt.show()
```
